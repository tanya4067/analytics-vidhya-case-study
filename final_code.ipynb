```python
def date_value(s):
    if(s=='0'):
        return(0)
    elif('-' in s):
        return(pd.to_datetime(s))
    else:
        return(pd.to_datetime(int(s)))
    
        
```

```python
s='U106593'
c=1
d=dict()
d1=dict()
arr=[]
count=0
for i in range(650695):
    t=merge['UserID'][i]
    if(t==s):
        if(merge['day'][i]>=13 and merge['day'][i]<=27):
            t1=(str(merge['Activity'][i])).upper()
            if(t1=='PAGELOAD'):
                p=merge['ProductID'][i]
                if(p!=0):
                    if(p not in d):
                        d[p]=c
                        d1[p]=merge['VisitDateTime'][i]
                        
                    else:
                        d[p]=d[p]+1
                        d1[p]=str(merge['VisitDateTime'][i])
    else:
        if(len(d)==0):
            sample['Most_Viewed_product_15_Days'][count]='Product101'
            
        else:
            m=0
            for i,j in d.items():
                if(m<j):
                    m=j
                    k=i
                elif(m==j):
                    m=j
                    f1=date_value(d1[k])
                    f2=date_value(d1[i])
                    if((f1<f2)==True):
                        k=i
                    else:
                        k=k
        sample['Most_Viewed_product_15_Days'][count]=k
        count+=1
        s=t
        d=dict()
        d1=dict()
sample['Most_Viewed_product_15_Days'][count]=k
```
